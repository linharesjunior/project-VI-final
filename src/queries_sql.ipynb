```python
import pandas as pd
from sqlalchemy import create_engine
from pandasql import sqldf
```

```python
layoff = pd.read_csv('../data/layofss.csv', sep=';')
layoff
```

```python
# Define the pandasql object
pysql = sqldf("SELECT * FROM layoff", globals())
pysql
```

```python
# Run SQL queries from companies that had laid_off more than 1000 employees 
result = sqldf("SELECT * FROM layoff WHERE laid_off > 1000")
result
```

```python
# Selecting rows where the percentage is greater than 10%

sqldf('''SELECT * 
FROM layoff 
WHERE percentage > 10''')
```

```python
# Grouping by industry and calculating the average percentage

sqldf('''SELECT industry, AVG(percentage) as avg_percentage 
FROM layoff 
GROUP BY industry''')
```

```python
# Sorting by date in descending order

sqldf('''SELECT * 
FROM layoff 
ORDER BY date DESC''')
```

```python
sqldf('''SELECT company, industry 
FROM layoff 
LIMIT 5''')
```

```python
sqldf('''SELECT company, date, industry
FROM layoff
ORDER BY date
LIMIT 10''')
```

```python
# Identify the companies that laid off the highest percentage of employees.

sqldf('''SELECT company, MAX(percentage) AS max_percentage
FROM layoff
GROUP BY company
ORDER BY max_percentage DESC
LIMIT 10''')
```

```python
# Analyze the distribution of layoffs across different industries

sqldf('''SELECT industry, COUNT(*) AS num_layoffs
FROM layoff
GROUP BY industry
ORDER BY num_layoffs DESC''')
```

```python
# Investigate the relationship between funding raised and layoffs

sqldf('''SELECT company, "raised(mm)" AS raised_amount, AVG(percentage) AS avg_percentage
FROM layoff
GROUP BY company
ORDER BY raised_amount DESC
LIMIT 10''')
```

```python
# Explore the timing of layoffs in different countries

sqldf('''SELECT country, COUNT(*) AS num_layoffs
FROM layoff
GROUP BY country
ORDER BY num_layoffs DESC''')
```

```python
# Examine the stage of companies that experienced layoffs

sqldf('''SELECT stage, AVG(percentage) AS avg_percentage
FROM layoff
GROUP BY stage
ORDER BY avg_percentage DESC''')
```
