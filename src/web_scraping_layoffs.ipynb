```python
import pandas as pd
import pdfplumber as pp
```

```python
# From website https://layoffs.fyi/, downloaded the layoff list as PDF
# Transform it to CSV indo a dataframe

with pp.open('../data/pdf.pdf') as pdf:
    df_final = pd.DataFrame()
    for n in range(len(pdf.pages)):
        table = pdf.pages[n].extract_table()

        header = table[0]
        dt = table[1:]

        df_page = pd.DataFrame(dt, columns=header)

        df_page['Location HQ'] = df_page['Location HQ'].str.replace('\\n', ', ', regex=True)
        
        df_final = pd.concat([df_final,  df_page], ignore_index=True)
```

```python
df_final.sample()
```

```python
df_final.rename(lambda x: x.lower().replace(' ', '_'), axis=1, inplace=True)
```

```python
df_final.rename(columns={'%': 'percentage'}, inplace=True)
df_final.rename(columns={'#_laid_off': 'laid_off'}, inplace=True)
df_final.rename(columns={'$_raised_(mm)': 'raised(mm)'}, inplace=True)
```

```python
df_final.sample()
```
