```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import os
import time
from dotenv import load_dotenv
load_dotenv()

driver = webdriver.Chrome()

# Increase the implicit wait time to 20 seconds
driver.implicitly_wait(20)

# Navigate to LinkedIn
driver.get("https://www.linkedin.com")

# Locate and fill in the login fields
username = driver.find_element(By.ID, "session_key")
password_input = driver.find_element(By.ID, "session_password")

# Retrieve email and password from environment variables
email = os.getenv("email")
password = os.getenv("password")

username.send_keys(email)
password_input.send_keys(password)
password_input.send_keys(Keys.RETURN)

time.sleep(2)

# Wait for the search bar to be present on the page
search_bar = driver.find_element(By.XPATH, "/html/body/div[6]/header/div/div/div/div[1]/input")

# Clear any existing text in the search bar
search_bar.clear()

# Search for "#layoff"
search_bar.send_keys("#layoff")
search_bar.send_keys(Keys.RETURN)

time.sleep(5)

# Select the filter Posts
try:
    posts_filter = driver.find_element(By.XPATH, "//button[text()='Posts']")
    posts_filter.click()
except Exception as e:
    print(f"Error clicking on the 'Posts' filter: {e}")
    
# Scroll down the page to load more content
for _ in range(3):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    
# Extract data from posts
post_data = []
try:
    posts = driver.find_elements(By.XPATH, "//li[contains(@class, 'search-result__occluded-item')]")
    for post in posts:
        post_text = post.text
        post_data.append({"Post Text": post_text})
except Exception as e:
    print(f"Error extracting post data: {e}")

# Create a DataFrame
linkedin = pd.DataFrame(post_data)
```

```python
url = 'https://www.linkedin.com'
driver = webdriver.Chrome()
driver.get(url)

time.sleep(2)

email_box = driver.find_element(By.ID, "session_key")
password_box = driver.find_element(By.ID, "session_password")

# 2. Declare variables
email = os.getenv("email")
password = os.getenv("password")

# 3. Send the text
email_box.send_keys(email)
password_box.send_keys(password)

# 4. Hit enter
password_box.send_keys(Keys.RETURN)

time.sleep(20)
# Navigate to jobs tab
jobs_button = driver.find_element(By.XPATH, '/html/body/div[6]/header/div/nav/ul/li[3]/a')
jobs_button.click()

time.sleep(2)
# Filter for job title
job_box = driver.find_element(By.XPATH, '/html/body/div[6]/header/div/div/div/div[2]/div[2]/div/div/input[1]')
time.sleep(1)
job_box.send_keys('deep learning engineer')
job_box.send_keys(Keys.ENTER)

time.sleep(2)
# Filter for location
location_box = driver.find_element(By.XPATH, '/html/body/div[6]/header/div/div/div/div[2]/div[2]/div/div/input[1]')
time.sleep(1)
location_box.clear()
time.sleep(3)
location_box.send_keys('Estados Unidos')
location_box.send_keys(Keys.ENTER)

# Save current link
current_link = driver.current_url

def write_file(el):
    job_id = el.get('data-job-id')

    # Open the web page
    driver.get(url + el.find('a').get('href'))
    time.sleep(3)

    html = driver.execute_script("return document.body.outerHTML;") # Javascript script
    soup = BeautifulSoup(html, 'html.parser')

    file_path = f'../data/jobs_html/{job_id}.txt'
    with open(file_path, 'w') as file:
        file.write(str(soup))
    time.sleep(3)

def create_html_files(list_):
    for i in list_:
        write_file(i)

# Job scrapping with bs4
html = driver.execute_script("return document.body.outerHTML;")
soup = BeautifulSoup(html, 'html.parser')

pagination = soup.find_all('ul', {'class': 'artdeco-pagination__pages--number'})
pagination = pagination[0].find_all('button')[:8]

for btn in pagination:
    for i in range(5):  # You can adjust the range as needed
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    aria = btn.get("aria-label")
    button = driver.find_element(By.XPATH, f'//button[@aria-label="{aria}"]')
    button.click()

    current_link = driver.current_url
    html = driver.execute_script("return document.body.outerHTML;")
    soup = BeautifulSoup(html, 'html.parser')

    time.sleep(3)
    container = soup.find_all('ul', {'class': 'scaffold-layout__list-container'})
    container = container[0]
    list_ = container.find_all('div', {'class': 'job-card-container'})
    create_html_files(list_)
    driver.get(current_link)
    container = []
    list_ = []
    time.sleep(3)

driver.quit()
```
